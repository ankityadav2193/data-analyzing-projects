#Uber Trips Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv('/uber-raw-data-sep14.csv')
```

```python
data.head()
```

```python
data['Date/Time']=data['Date/Time'].map(pd.to_datetime)
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.describe()
```

```python
data['Day']=data['Date/Time'].apply(lambda x:x.day)
```

```python
data['Month']=data['Date/Time'].apply(lambda x:x.month)
```

```python
data['Year']=data['Date/Time'].apply(lambda x:x.year)
```

```python
data['Hour']=data['Date/Time'].apply(lambda x:x.hour)
```

```python
data['Minute']=data['Date/Time'].apply(lambda x:x.minute)
```

```python
data['weekday']=data['Date/Time'].apply(lambda x:x.weekday())
```

```python
data.head()
```

```python
data.shape
```

```python

sns.distplot(data["Day"])
```

```python
sns.distplot(data["Month"])
```

```python
sns.distplot(data["Year"])
```

```python
sns.distplot(data["Hour"])
```

```python
sns.distplot(data["weekday"])
```

```python
sns.distplot(data["Minute"])
```

```python
# Correlation of Weekday and Hour
df = data.groupby(["weekday", "Hour"]).apply(lambda x: len(x))
df = df.unstack()
sns.heatmap(df, annot=False)
```

```python
data.plot(kind='scatter', x='Lon', y='Lat', alpha=0.4, s=data['Day'], label='Uber Trips',
figsize=(12, 8), cmap=plt.get_cmap('jet'))
plt.title("Uber Trips Analysis")
plt.legend()
plt.show()
```
